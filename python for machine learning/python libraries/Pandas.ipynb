# Marge
# primary key, foreign key diye 2 table er data concat/ Marge kora

```
left = pd.DataFrame({'key': ['K3', 'K2', 'K1', 'K0'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})

print(left)
print()
print(right)
```

```

pd.merge(left , right , how = 'inner' , on = 'key')
```

# how{‘left’, ‘right’, ‘outer’, ‘inner’}
, default ‘inner’<br>
Type of merge to be performed.
<br>
left: use only keys from left frame, similar to a SQL left outer join; preserve key order.
<br>
right: use only keys from right frame, similar to a SQL right outer join; preserve key order.
<br>
outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.
<br>
inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.<br>

```

left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print()
print(right)
```

```
# marge with 2 keys:

pd.merge(left,right , on = ['key1' , 'key2'])
```

```
pd.merge(left,right ,how='outer', on = ['key1' , 'key2'])
```

```
pd.merge(left,right , how = 'right' , on = ['key1' , 'key2'])
```

# <strong>join operation</strong>

```
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
print(left)
print()
print(right)
```

```
left.join(right) # here priority is left 
```

```
left.join(right , how = 'right')  # here priority is  right
```

```
left.join(right , how = 'outer') # take all
```
